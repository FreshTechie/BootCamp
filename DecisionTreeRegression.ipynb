```python
# -*- coding: utf-8 -*-
"""
Created on Fri Oct 19 19:47:25 2018

@author: Shiva Thiru
"""

#Imported Required Libraries

import numpy as np
import pandas as pd
from sklearn.cross_validation import train_test_split 
from sklearn.tree import DecisionTreeRegressor  
from sklearn.metrics import accuracy_score 
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

#Building Phase

# Loading Required columns from csv files into Pandas Dataframe for Data Cleaning and Processing
train_data =pd.read_csv('interview-dataset.csv'
                , names=['pred_maint_data_flg','block','batt_volt','comp_type','cur_after300ms_c','cur_after300ms_o','cur_c_arr.38','cur_o_arr.33','datetime','direction','door','door_ref.DR5','door_ref.equip_loc','eventcounter','index','kc_after300ms_c','kc_after300ms_o','loc_door_c.openswtch','loc_door_c.swtch2','loc_door_o.openswtch','loc_door_o.swtch2','maint_ver','max_calcstroke_c','message','numcumulsecpoweron','numunlockpoweron','obstacledetected_c','obstacledetected_o','product_type','pwm_after300ms_c','pwm_after300ms_o','residue_fcv','resist_after300ms_c','resist_after300ms_o','revision','software_ver.octet3','station','time_first_opening','time_last_closing','train_id','ttime_c_arr.38','ttime_c_spd_arr.38','ttime_o_arr.33','ttime_o_spd_arr.33','volt_after300ms_c','volt_after300ms_o','volt_o_arr.20'])
  
# Converting all datatype as int,so that we can fit the data in Regression Fit Method
 
train_data['pred_maint_data_flg'],_ = pd.factorize(train_data['pred_maint_data_flg'])
train_data['block'],_ = pd.factorize(train_data['block'])
train_data['batt_volt'],_ = pd.factorize(train_data['batt_volt'])
train_data['comp_type'],_ = pd.factorize(train_data['comp_type'])                   
train_data['cur_after300ms_c'],_ = pd.factorize(train_data['cur_after300ms_c'])
train_data['cur_after300ms_o'],_ = pd.factorize(train_data['cur_after300ms_o'])
train_data['cur_c_arr.38'],_ = pd.factorize(train_data['cur_c_arr.38'])
train_data['cur_o_arr.33'],_ = pd.factorize(train_data['cur_o_arr.33'])
train_data['datetime'],_ = pd.factorize(train_data['datetime'])
train_data['direction'],_ = pd.factorize(train_data['direction'])  
train_data['door'],_ = pd.factorize(train_data['door'])  
train_data['door_ref.DR5'],_ = pd.factorize(train_data['door_ref.DR5'])  
train_data['door_ref.equip_loc'],_ = pd.factorize(train_data['door_ref.equip_loc'])  
train_data['eventcounter'],_ = pd.factorize(train_data['eventcounter'])
train_data['index'],_ = pd.factorize(train_data['index'])
train_data['kc_after300ms_c'],_ = pd.factorize(train_data['kc_after300ms_c'])
train_data['kc_after300ms_o'],_ = pd.factorize(train_data['kc_after300ms_o'])
  
train_data['loc_door_c.openswtch'],_ = pd.factorize(train_data['loc_door_c.openswtch'])
train_data['loc_door_c.swtch2'],_ = pd.factorize(train_data['loc_door_c.swtch2'])
train_data['loc_door_o.openswtch'],_ = pd.factorize(train_data['loc_door_o.openswtch']) 
train_data['loc_door_o.swtch2'],_ = pd.factorize(train_data['loc_door_o.swtch2'])
train_data['maint_ver'],_ = pd.factorize(train_data['maint_ver'])
train_data['max_calcstroke_c'],_ = pd.factorize(train_data['max_calcstroke_c'])
train_data['message'],_ = pd.factorize(train_data['message'])
train_data['numcumulsecpoweron'],_ = pd.factorize(train_data['numcumulsecpoweron'])
train_data['numunlockpoweron'],_ = pd.factorize(train_data['numunlockpoweron'])
train_data['obstacledetected_c'],_ = pd.factorize(train_data['obstacledetected_c'])
train_data['obstacledetected_o'],_ = pd.factorize(train_data['obstacledetected_o'])

train_data['product_type'],_ = pd.factorize(train_data['product_type'])
train_data['pwm_after300ms_c'],_ = pd.factorize(train_data['pwm_after300ms_c'])
train_data['pwm_after300ms_o'],_ = pd.factorize(train_data['pwm_after300ms_o'])
train_data['residue_fcv'],_ = pd.factorize(train_data['residue_fcv'])

train_data['resist_after300ms_c'],_ = pd.factorize(train_data['resist_after300ms_c'])
train_data['resist_after300ms_o'],_ = pd.factorize(train_data['resist_after300ms_o'])
train_data['revision'],_ = pd.factorize(train_data['revision'])  
train_data['software_ver.octet3'],_ = pd.factorize(train_data['software_ver.octet3'])
train_data['station'],_ = pd.factorize(train_data['station'])
train_data['time_first_opening'],_ = pd.factorize(train_data['time_first_opening'])
train_data['time_last_closing'],_ = pd.factorize(train_data['time_last_closing']) 
train_data['train_id'],_ = pd.factorize(train_data['train_id'])
                

train_data['volt_o_arr.20'],_ = pd.factorize(train_data['volt_o_arr.20'])
train_data['volt_after300ms_o'],_ = pd.factorize(train_data['volt_after300ms_o'])
train_data['volt_after300ms_c'],_ = pd.factorize(train_data['volt_after300ms_c'])
train_data['ttime_o_spd_arr.33'],_ = pd.factorize(train_data['ttime_o_spd_arr.33'])
train_data['ttime_o_arr.33'],_ = pd.factorize(train_data['ttime_o_arr.33'])
train_data['ttime_c_spd_arr.38'],_ = pd.factorize(train_data['ttime_c_spd_arr.38'])
train_data['ttime_c_arr.38'],_ = pd.factorize(train_data['ttime_c_arr.38'])

# Full Details of Entrie DataSet
#train_data.info()

# Taken column 1 to 46 as Features for this Prediction    
X = train_data.values[:,1:46]
# Taken column pred_maint_data_flg as Label for this Prediction
Y = train_data.pred_maint_data_flg

#Fitting both Features and Labels in Train_Test_Split Method    
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 100) 
     

# Fitting Training Dataset into Decision Tree Regressor Fit Method
regressor = DecisionTreeRegressor()  
regressor.fit(X_train, y_train)  
   
    
# Operational Phase 
    
# Once Model trained with Train DataSet, Predicted Test DataSet using Regressor Predict Method
y_pred_gini = regressor.predict(X_test) 
#print("y_pred_gini",y_pred_gini)


#print("Total TestDataSet considered for Prediction-->",(y_pred_gini))
count_total = (y_test).shape
print('Total Count: {}'.format(count_total))

  
#Total Abnormal Test DataSet Count
count_mismatch = (y_test != y_pred_gini).sum()
print('Abnormal Count: {}'.format(count_mismatch))

#Total Normal Test DataSet Count
count_match = (y_test == y_pred_gini).sum()
print('Normal Count: {}'.format(count_match))

#Printing Accuracy for the Prediction 
accuracy=accuracy_score(y_test, y_pred_gini)
print('Accuracy: {:.2f}'.format(accuracy))

#Printing Error Details for Prediction  
print('Mean Absolute Error:', mean_absolute_error(y_test, y_pred_gini))  
print('Mean Squared Error:', mean_squared_error(y_test, y_pred_gini))  
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test, y_pred_gini)))  

#Printing Normal and AbNormal Samples after Prediction and Abnormal Samples required Train Door Maintanence
print("Abnormal Samples-->",(y_test != y_pred_gini))
print("Normal Samples-->",(y_test == y_pred_gini))

```
